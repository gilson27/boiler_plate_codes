```
import threading
```

```
shared_data = 40
lock = threading.Lock()

#ROS
#MQTT
#GIL reactive architecture

def find_sum():
  print("In Find Sum")
  global shared_data
  lock.acquire()
  shared_data = shared_data + 10
  print("find_sum" + str(shared_data))
  lock.release()

def find_product():
  print("In Find Product")
  global shared_data
  lock.acquire()
  shared_data = shared_data * 2
  print("find_product" + str(shared_data))
  lock.release()

def main():
  th1 = threading.Thread(target=find_sum)
  th2 = threading.Thread(target=find_product)

  th1.start()
  th2.start()

  th1.join()
  th2.join()



```

```
main()
```
